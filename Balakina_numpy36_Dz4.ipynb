```python
# Работа с словоформами - русский - все форма к им п ед.ч.
from pymystem3 import Mystem
m=Mystem()
```

```python
def get_lemmas(keyword):
  list = keyword.lower().strip().split() # разбиваем каждую строку keywords на отдельные слова и проверяем  в них
  result=[]
  for element in list:
    result.append(m.lemmatize(element))
  return result  
```

```python
keywords['lemmas'] = keywords['keyword'].apply(get_lemmas) # считалось почти 10 часов!!! и не сохранилось!
```

## Вариант 1 . Решение с помощью user-fefined функции и .apply() - самый медленный

```python
def fill_sourse_type(row):
  """создайте столбец source_type по следующим правилам:

  если источник traffic_source равен yandex или google, 
  то в source_type ставится organic;
  для источников paid и email из России 
  ставим ad,
  для источников paid и email не из России  
  ставим other, 
  все остальные варианты берем из traffic_source без изменений """
  if row['traffic_source'] in['yandex', 'google']:
      return 'organic'
  elif (row['traffic_source'] in ['paid', 'email']) and (row['region'] =='Russia'):
      return 'ad'
  elif (row['traffic_source'] in ['paid', 'email']) and (row['region']!='Russia'):
      return 'other'
  else:
      return row['traffic_source']
```

```python
%%time
log['source_type'] = log.apply(fill_sourse_type, axis = 1) # заносим в столбец source_type значения, полученные с помощью функции Fill_source_type на основе данных столбца 'traffic_source'(берем из строки row)
```

```python
log['source_type'].value_counts()
```

##Вариант 2. Решение с помощью lambda, map, самое быстрое выходит в 25 раз быстрее по сравнению с вариантом 1!

```python
%%time
log['source_type'] = list(map(lambda x,y:'organic' if x in ['yandex', 'google'] else ('ad' if x in  ['paid', 'email'] and y =='Russia' else ('other' if x in  ['paid', 'email'] and y !='Russia' else x )) , log['traffic_source'], log['region']))
```

```python
log['source_type'].value_counts()
```

## Вариант 3. Решение с использованием loc и str.contains(), медленнее в 4 раза по сравнению с вариантом 2

```python
%%time
log.loc[log.traffic_source.str.contains('yandex|google'), 'source_type'] ='organic'
log.loc[(log.traffic_source.str.contains('paid|email')) & (log.region =='Russia'), 'source_type'] ='ad'
log.loc[(log.traffic_source.str.contains('paid|email')) & (log.region !='Russia'), 'source_type'] ='other'
log.source_type.fillna(value = log.traffic_source, inplace = True)
```

```python
log['source_type'].value_counts()
```

## Вариант 4. Решение с использованием loc и isin (Совет преподавателя). Немного медленнее чем вариант 2, но быстрее, чем 1 и 3

```python
%%time
log.loc[log.traffic_source.isin(['yandex', 'google']), 'source_type'] = 'organic'
log.loc[(log.traffic_source.isin(['paid', 'email'])) & (log.region =='Russia'), 'source_type'] ='ad'
log.loc[(log.traffic_source.isin(['paid', 'email'])) & (log.region !='Russia'), 'source_type'] ='other'
log.source_type.fillna(value = log.traffic_source, inplace = True)
```

# Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
data_ratings = pd.read_csv('/content/ml-latest-small/ratings.csv') 
```

```python
data_ratings.head()
```

## Вариант 1. Решение с помощью  lambda

```python
%%time
data = data_ratings.groupby('userId').agg({'timestamp': ['min', 'max', 'count']})
data = data[data['timestamp']['count']>100]
data['lifetime'] = data.apply(lambda row: row['timestamp']['max']-row['timestamp']['min'], axis = 1 )
print(f' среднее время жизни пользователей, которые выставили более 100 оценок: {data.lifetime.mean()/24/3600}')
```

## Вариант 2. Решение с помощью data.loc - быстрее в 4 раза - совет преподавателя

```python
%%time
data = data_ratings.groupby('userId').agg({'timestamp': ['min', 'max', 'count']}) # Здесь -минимальное и максимальное timestamp, а также кол-во оценокэ. Вс- внутри timestamp
data
```

```python
data = data.loc[data['timestamp']['count']>100]
data['lifetime'] = data['timestamp']['max']-data['timestamp']['min']
print(f'среднее время жизни пользователей, которые выставили более 100 оценок: {data.lifetime.mean()/24/3600}')
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```
