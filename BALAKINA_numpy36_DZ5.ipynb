## Обязательная часть
Вам необходимо написать функцию, которая будет основана на поиске по сайту habr.com. Функция в качестве параметра должна принимать список запросов для поиска (например, ['python', 'анализ данных']) и на основе материалов, попавших в результаты поиска по каждому запросу, возвращать датафрейм вида:

<дата> - <заголовок> - <ссылка на материал>
В рамках задания предполагается работа только с одной (первой) страницей результатов поисковой выдачи для каждого запроса. Материалы в датафрейме не должны дублироваться, если они попадали в результаты поиска для нескольких запросов из списка.

```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup
import time
```

```python
def get_habr_blog_posts(list_of_queries):
    """
    функция основана на поиске по сайту habr.com. 
    Параметр - список запросов для поиска (например, ['python', 'анализ данных']) 
    Результат - датафрейм вида:
    <дата> - <заголовок> - <ссылка на материал>, 
    содержащий 1-ю страницу поисковой выдачи.
    материалы датафрейма не должны дублироваться
  """
    params = dict()
    url_domain = 'https://habr.com'
    url_search = url_domain + '/ru/search/'
    
    for query in list_of_queries:

        params = {'q': query }  
        req = rq.get(url_search, params).text
        soup = BeautifulSoup(req)
        news = soup.find_all('article', class_ ='tm-articles-list__item')
        habr_blog = pd.DataFrame(columns = ['title', 'date', 'link'])

        for el in news:
            if el.find('h2', class_ = 'tm-article-snippet__title tm-article-snippet__title_h2'):
                title_ = el.find('h2', class_ = 'tm-article-snippet__title tm-article-snippet__title_h2').text
            else:
                title_ = el.find('h2', class_ = 'tm-megapost-snippet__title').text

            if el.find('a', class_ ='tm-article-snippet__title-link'):        
                link_ = url_domain + el.find('a', class_ ='tm-article-snippet__title-link').get('href')
            else:
                link_ =  url_domain + el.find('a', class_ = 'tm-megapost-snippet__link tm-megapost-snippet__card').get('href')

            if el.find('time'):   
                date_ = el.find('time').text
            row_ = {'date': date_, 'title': title_, 'link': link_}
            habr_blog = pd.concat([habr_blog, pd.DataFrame([row_])]) 
            
          
    return habr_blog.reset_index(drop=True)
```

# Дополнительная часть (необязательная)
Функция из обязательной части задания должна быть расширена следующим образом:

кроме списка ключевых слов для поиска необходимо объявить параметр с количеством страниц поисковой выдачи. Т.е. при передаче в функцию аргумента 4 необходимо получить материалы с первых 4 страниц результатов;
в датафрейме должны быть столбцы с полным текстом найденных материалов и количеством лайков:
<дата> - <заголовок> - <ссылка на материал> - <текст материала> - <количество лайков>

```python
def get_list_of_query():
  """
  Формирует список из вводимых с клавиатуры построчно поисковых запросов
  Запросы добавляются в список, пока не введено 'q'
  """
  list_of_query =[]
  query = ''
  while query != 'q':
    query = input('Введите запрос или "q" : ')
    list_of_query.append(query)
  return list_of_query[:-1]
```

```python

def get_habr_blog_posts(list_of_queries, pages):
    """
функция основана на поиске по сайту habr.com. 
Параметры - list_of_queries -список запросов для поиска
 (например, ['python', 'анализ данных']) и
            pages -количество страниц поисковой выдачи
Результат - датафрейм вида:
<дата> - <заголовок> - <ссылка на материал>-<текст материала> - <количество лайков>, 
материалы датафрейма не должны дублироваться
  """
    params=dict()
    url_domain='https://habr.com'

    for query in list_of_queries:

        url_search = url_domain + '/ru/search/'
        params = {'q': query }
        req = rq.get(url_search, params).text # отправляем запрос на страницу поиска с передачей параметра поиска - элемента из списка запросов list_of_queries

        time.sleep(0.3)

        soup = BeautifulSoup(req) # получаем объект с поисковой выдачей по 1-й странице, из которого получим данные по содержимому пагинатора
        paginator_contents = soup.find_all('a', class_ = 'tm-pagination__page') # получаем объект с содержимым пагинатора 



        all_pages_numbers =[] # инициируем список номеров страниц из пагинатора
        all_pages_links =[] # инициируем список ссылкок на страницы поисковой выдачи

        for element in paginator_contents:
          all_pages_numbers.append(element.text.strip()) # добавляем номера страниц из пагинатора в список. Наи важно, чте здесь есть номер последней страницы
         
       
        max_number = all_pages_numbers[-1] # Это номер последней сраницы
        if max_number < str(pages): # Проверяем, не слишком ли много  страниц поисковой выдачи  мы запросили
          pages = int(max_number) # в любом случае будут выведены только запросы со страниц поисковой выдачи из пашгинатора

        for i in range(1, pages+1):
           all_pages_links.append (url_search +'page'+ str(i)+'/')

      
        habr_blog = pd.DataFrame(columns = ['title', 'date', 'link','text', 'likes'])
        for page_link in set(all_pages_links):  # поочередно проходим все страницы поисковой выдачи из заданного кол-ва страницобеспечиваем уникальность поисковых ссылок
              
              req = rq.get(page_link, params).text
              time.sleep(0.3)
              soup = BeautifulSoup(req) # А здесь у же - с поисковой выдачей по текущей странице
              news = soup.find_all('article', class_ ='tm-articles-list__item')
             

              for el in news:

                  if el.find('h2', class_ = 'tm-article-snippet__title tm-article-snippet__title_h2'):
                      title_ = el.find('h2', class_ = 'tm-article-snippet__title tm-article-snippet__title_h2').text
                  else:
                      title_ = el.find('h2', class_ = 'tm-megapost-snippet__title').text

                  if el.find('a', class_ ='tm-article-snippet__title-link'):        
                      link_ = url_domain + el.find('a', class_ ='tm-article-snippet__title-link').get('href')
                  else:
                      link_ =  url_domain + el.find('a', class_ = 'tm-megapost-snippet__link tm-megapost-snippet__card').get('href')

                  if el.find('time'):   
                      date_ = el.find('time').text
                  
                  if el.find('div', class_ = 'tm-votes-meter tm-data-icons__item' ):        
                      likes_ = el.find('div', class_ = 'tm-votes-meter tm-data-icons__item').text
                  else:
                      likes_ = el.find('div', class_ = 'tm-votes-meter tm-data-icons__item').text

                  res = rq.get(link_).text
                  time.sleep(0.3)
                  soup_ = BeautifulSoup(res)
                  
                  if  soup_.find('div', class_ ='tm-article-body'):
                    text_ = soup_.find('div', class_ ='tm-article-body').text.strip()

                  row_ = {'date': date_, 'title': title_, 'link': link_, 'text': text_, 'likes': likes_ }
                  habr_blog = pd.concat([habr_blog, pd.DataFrame([row_])]) 
               
      
    return habr_blog.reset_index(drop=True)
```
