## Вариант 1 - с использованием user fefined функции classificator

```python
%%time
def classificator(rating):
  """присваивает классификатор в зависимости от рейтинга"""
  if rating <=2:
    return 'низкий рейтинг' 
  elif rating <=4:
    return'средний рейтинг'
  else:
    return 'высокий рейтинг'
data_ratings['class'] = data_ratings.rating.apply(classificator) # применяем ф-ю classificator для расчета значения столбца class
data_ratings.head()
```

## Вариант 2 - с использованием lambda и map(), немного дольше

```python
%%time
data_ratings['class'] = list(map(lambda x: 'низкий рейтинг' if x <= 2 else('средний рейтинг' if x <= 4 else 'высокий рейтинг'),data_ratings['rating']))
data_ratings.head()
```

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

В переменную years запишите список из всех годов с 1950 по 2010.

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

для каждой строки пройдите по всем годам списка years
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
data_ratings = pd.read_csv('/content/ml-latest-small/ratings.csv') # Читаем файл raiting.csv
```

```python
with ZipFile('ml-latest-small.zip', 'r') as zip_file:  # Распаковываем архив в сессионное хранилище в папку ml-latest-small. Нам понадобятся файлы ratings.csv и movies.csv
     #zip_file.extractall('/content/')
     zip_file.extract('ml-latest-small/movies.csv')
```

```python
data_movies = pd.read_csv('/content/ml-latest-small/movies.csv') # теперь читаем файл movies.csv с названиями фильмов
data_movies.head()
```

```python
def production_year(title):
  """каждой строке из названия фильма выставляет год выпуска"""
  years = list(range(1950,2011))
  for year in years:
    if str(year) in title:
      return str(year)
  return '1900'
```

```python
data_movies['year'] = data_movies.title.apply(production_year) # Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’
data_movies.head()
```

```python
joined = data_ratings.merge(data_movies, on ='movieId', how = 'left') # соединяем оба датафрейма
```

```python
joined[['year','rating']].groupby('year').agg('mean').sort_values(by = 'rating', ascending = False) # оставляем столбцы 'year' и 'rating', группируем по годам, считаем средний рейтинг, сортируем по убыванию рейтинга

```
